```python
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on='title')
movies.shape
```

```python
movies = movies[['id','title','genres','keywords','overview','cast','crew']]
```

```python
movies.head()
```

```python
movies['genres'][0]
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    l = []
    obj = ast.literal_eval(obj)
    for i in obj:
        l.append(i['name'])
    return l
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    l = []
    c = 1
    obj = ast.literal_eval(obj)
    for i in obj:
        if c<=3:
            l.append(i['name'])
            c+=1
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def convert2(obj):
    l = []
    obj = ast.literal_eval(obj)
    for i in obj:
        if i['job']=='Director':
            l.append(i['name'])
    return l
```

```python
convert2(movies['crew'][0])
```

```python
movies['crew'] = movies['crew'].apply(convert2)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
df = movies[['id','title','tags']]
```

```python
df['tags'] = df['tags'].apply(lambda x:' '.join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x: x.lower())
```

```python
import nltk
```

```python
import re
```

```python
steammer = nltk.stem.PorterStemmer()
```

```python
stopwords=nltk.corpus.stopwords.words('english') 
```

```python
def processes(r1):
    r1 = r1.lower()
    r1 = re.sub('[^a-z]',' ',r1)
    words = [steammer.stem(i) for i in r1.split() if i not in stopwords]
    return ' '.join(words)
```

```python
df['tags'] = df['tags'].apply(processes)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
vectorizer =  CountVectorizer(max_features=5000)
bow = vectorizer.fit_transform(df['tags'])
x_train = bow.toarray()
```

```python
x_train.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(x_train)
```
